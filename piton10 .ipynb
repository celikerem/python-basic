# If-Elif-Else Conditional Statements (Şart durumları)

```python
# If 'eğer' anlamına gelir. Devamında gelen koşulun sağlanması durumunda kullanılır.
# elif(else if) 'eğer değilse' anlamını taşır. if kodu gerçekleşmezse bu şarta bakılır bu şartın sağlanması durumunda çalışır.
# else 'değilse' anlamını taşır. İf ve elif anlamının hiçbiri gerçekleşmezse bu şart çalışır.
```

```python
a = 1

if True:
  print(a)
```

```python
if False:
  print(a)
```

```python
if a<5:
  print("burası çalışacak")
```

```python
b = 10

if b >= 20:
  print(f"{b} sayısı 20 den büyüktür")
else:
  print("üstteki şartlar doğru değilmiş")
```

```python
c = 15

if c < 20:
  print(f"{c} sayısı 20den küçükmüş")
elif c < 50:
  print(f"{c} sayısı 50'den küçükmüş")
else:
  print("üstteki şartların hiçbiri geçerli değil")
```

```python
# kullanıcıdan yaş bilgisini alalım yaşı 18den küçükse 'ehliyet almak için uygun değilsiniz, eğer yaşı 18den büyükse "ehliyet alabilirsiniz"
yas =int(input("yaşınız"))
if yas<18:
  print("ehliyet almak için uygun değilsiniz")
elif yas>=18:
  print("ehliyet almak için uygunsunuz")
```

```python
#soru kullanıcıdan bir şifre bilgisi alalım şifre "bilisim" ise "giriş başarılıiyönlendiriliyorsunuz " yazsın şifre başka bir şey işe 'şifreniz yanlış'
sifre=input("şifeniz :")
if sifre=='bilisim':
  print("giriş başarılı, yönlendiriliyorsunuz")
else:
  print("şifreniz yanlış")
```

```python
#kullanıcıdan 0-100 arasında bir sayı alalım. alınan sayı 1 basamaklı ise "bir basamaklıdır" alınan sayı 2 basamaklı ise "2 basamaklı"
sayii=int(input("0 ile 100 arasında bir tam sayı"))
if (sayii/10)<1:
  print("sayınız bir basamaklı")
elif (sayii/10) > 1:
  print("sayınız 2 basamaklı")
else:
  print("yanlış giriş")
```

```python
#kullanıcıdan bir tam sayı bilgisi alalım sayının tek ya da çift olması durumuna göre "tektir" "çifttir"
v = int(input("bir tam sayı girin"))
if v%2 == 0:
  print("çifttir")
elif v%2 == 1:
  print("tektir")
```
