# listeye eleman ekleme

```python
# iki şekilde listeye yeni eleman eklenebilir.
#1) .append() methodu ile listenin SONUNA
#2) .insert() methodu ile listenin istediğiniz bir indexine istediğimiz bir elemanı ekleyebiliriz.
```

```python
liste = ['derya',7,'aslı','ali',9,'cenk','kemal']
```

```python
liste.append('sandalye')
liste
```

```python
liste.append("burak")
```

```python
#görev: listenin 2. indexine "ipek yazısını ekleyelim"
liste.insert(2, "ipek")
liste
```

```python
#soru yaş listesi= 13,15,18,26 listesinin sonuna 17 ekleyiniz
yas_listesi = [13,15,18,26]
yas_listesi.append(17)
yas_listesi
```

```python
#isim listesi=["cengiz","ali","murat","ece","dilan"] listesinin -3. indexine kerem ekleyin
isimListesi = ["cengiz","ali","murat","ece","dilan"]
isimListesi.insert(-3 , "kerem")
isimListesi
```

#listeden eleman silme

```python
#3 farklı şekilde yapılabilir
#1) .remove methoduyla silinebilir(1 parametre alır silinecek olan elemanın ismini)
#2) .pop methoduyla silinebilir (1 parametre alır silinecek olan elemanın index numarasınu alır. default -1)
#3) del komutuyla silinebilir. (del liste[3])listenin 3.elemanı
```

```python
hava_durumu = ['bulutlu',"yağmurlu","karlı","güneşli","parçalı bulutlu"]
hava_durumu.remove("güneşli")
hava_durumu
```

```python
hava_durumu.pop(2)
hava_durumu
```

```python
del hava_durumu[1]
hava_durumu
```
