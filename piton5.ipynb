#slicing

```python
#slicing işlemi dilimleme demektir
#string bir ifadenin istediğimiz bir dilimini alırız.
#slicing işleminde  ['start' ' stop' 'step'] adımları vardır
#slicing işleminde stop noktası dahil değildir.
#step yazılmazsa default olarak 1 kabul edilir
#stop noktası belirtilmezse sonuna kadar gider
```

```python
a = "kodlama eğitimi alıyorum"
```

```python
#slicing yaprak bu cümlenin içinden eğitimi çekiniz
```

```python
a[8:14]
```

```python
#soru: Aıo kısmını alın.
```

```python
a[16:21:2]
```

```python
a[-8:-3:2]
```

```python
#e den başlayıp sonuna kadar hepsini alalım
```

```python
a[8:]
```
