#formatlama

```python
#bir stringin içerisine önceden tanımlanan bir değişkeni göndermek için formatlama yapılır.
#formatlama için '{}' kullanılır.
```

```python
isim = 'Kerem'
soyisim = 'Çeliker'
yas = 14
okul = 'TOBB Fen Fisesi'
```

```python
#üstteki 3 değişkeni formatlama yaparak
#"merhabalar benim adım ..., soyadım ..., şuanda...yaşımdayım" yazdıralım
```

```python
#1. formatlama (en ilkel)

print("merhabalar benim adım" , isim , "soyadım" , soyisim , "yaşım" , yas)
```

```python
#2. formatlama (biraz daha yeni ve bütün programlama dillerinde benzer şekilde var)

print("merhabalar benim adım {}, soyadım {}, şuanda {} yaşıdayım".format(isim,soyisim,yas))
```

```python
#3. formatlama yöntemi (en iyisi ama sadece phytona özel. f-string denir)

print(f"merhabalar benim adım {isim}, soyadım {soyisim}, yaşım {yas}")
```

```python
#isim değişkenine isminizi soyisim değişkenine soyisminizi okul değşikenine okul değişkenini yazın devamında formatlama yaparak
#merhaba ben... ... şuanda ... okulunda okuyorum yazıdırn
```

```python
print(f"merhaba ben {isim} {soyisim}, şuanda {okul} okulunda okuyorum  ")
```

```python
#benim adım... babamın adı ... annemin adı... ve soyadımız ... 'dır yazdıralım
```
