```python
# bir tahmin oyunu oynayalaım. Bizim belirlediğimiz bir kelime olsun. kullanıcının bu kelimeyi tahmin etrmesi için 5 canı olsun
# kullanıcıdan kelime tahmlinlerini isteyeliam ne zaman bilirse '...' denemenizde doğru kelimeyi buldunuz.'
# kullanıcı 5 tahminde de doğru cevabı veremezse 'ne yazıkki kaybettiniz'

can=5
kelime='bilgisayar'

while can != 0:
  tahmin=input("tahmininizi giriniz")
  can-=1
  if tahmin==kelime:
    break
if can != 0:
  print(f'tebrikler {5-can}. denemede doğru kelimeyi buldunuz')
print("ne yazık ki kaybettiniz")
```

```python
#başka bir tahmin oyunu daha oynayalım. bu sefer bizim belirlediğimiz bir sayı olsun kullanıcı bu sayıyı tahmin etmeye çalışsın.
# kullanıcı daha kğçğk girerrse 'daha büyük bir sayı girin' kullanıcı büyük girerse ' daha küçük bir sayı girin'
# deneme sayısınıa göre çıktı

sayi=53
deneme=0

while True:
  tahmin=int(input("bir sayı tahmin edin"))
  deneme+=1
  if tahmin==sayi:
    print(f"{deneme}. denemede bildiniz")
    break
  elif tahmin>sayi:
    print(" daha küüçük bir sayı girmelisin")
  else:
    print("daha büyük bir sayı girmelisin")

```

```python
# kullanıcıdan 2 sayı isteyin 1. sayının 5'e tam bölünmesini 2. girdiğini 3e tam bölünmesini isteyin
#1. ve 2. sayı şartlara uymazsa ona göre çıktı verelim
#şartlara uyumlu giriş yaptığında ' tebrikler ....cı denemenizde doğru bildiniz çıktısı verelim

print("2 sayı girin 1.si 5e 2.si 3e tam bölünmeli")
deneme=0
while True:
  deneme+=1
  birincisayi=int(input("1. sayıyı girin 5e tam bölünen"))
  ikincisayi=int(input("2. sayıyı girin 3e tam bölünen"))
  if birincisayi%5!=0:
    print("girdiğiniz 1. sayı şartları sağlamıyor")
    continue
  if ikincisayi%3!=0:
    print("girdiğiniz 2. sayı şartları sağlamıyor")
    continue
  else:
    print(f"{deneme}. denemede doğru bildiniz")
    break
```
